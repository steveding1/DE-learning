```python
from ipaddress import ip_address
import pandas as pd
from http import HTTPStatus
import datetime
import sqlite3
```

```python
df = pd.read_csv('traffic.csv',parse_dates=['time'])
df.info()
```

```python
respond_code = [int(status) for status in set(HTTPStatus)]
median_time = df.time.median()
```

```python
def df_valid(row):
  try:
    ip_address(row['ip'])
  except ValueError:
    return False
  if abs(row['time'] - median_time) > datetime.timedelta(days=1):
    return False
  if pd.isnull(row['path']) or not row['path'].strip():
    return False
  if row['status'] not in respond_code:
    return False
  if pd.isnull(row['size']) or row['size']<=0:
    return False
  return True  
```

```python
bad_rows = df[~df.apply(df_valid,axis=1)]
bad_pct = len(bad_rows)/len(df)*100
print(bad_pct)
bad_rows
```

```python
df = df[~df.index.isin(bad_rows.index)]
df
```

```python
with sqlite3.connect('dbfile.db') as conn:
  df.to_sql('traffic',conn,if_exists='append',index=False)
```

```python
pd.read_sql('select * from traffic',conn)
```
